```python
import pandas as pd

# Membaca kedua file CSV
df1 = pd.read_csv('train2_subset.csv')
df2 = pd.read_csv('train_subset.csv')

# Menggabungkan kedua dataframe secara horizontal
df_combined = pd.concat([df1, df2], axis=1)

# Menghapus titik dari nilai dalam kolom 'price' dan mengubahnya menjadi integer
df_combined['price'] = df_combined['price'].str.replace('.', '').astype(int)

# Menyimpan dataframe gabungan ke dalam file CSV baru
df_combined.to_csv('file_gabungan.csv', index=False)
```

```python
df = pd.read_csv('file_gabungan.csv')

# Memastikan kolom harga dalam format numerik
df['price'] = pd.to_numeric(df['price'], errors='coerce')

# Memfilter data berdasarkan harga
filtered_df = df[df['price'] <= 8000000]

# Menampilkan data yang difilter
print(filtered_df)

# Menyimpan data yang difilter ke file CSV baru
filtered_df.to_csv('filtered_data_low_level.csv', index=False)
```

```python
import pandas as pd

# Membaca file CSV yang telah digabungkan
df = pd.read_csv('file_gabungan.csv')

# Memastikan kolom 'price' dalam format numerik
df['price'] = pd.to_numeric(df['price'], errors='coerce')

# Memfilter data berdasarkan harga
filtered_df = df[(df['price'] > 8000000) & (df['price'] <= 16000000)]

# Menampilkan data yang difilter
print(filtered_df)

# Menyimpan data yang difilter ke file CSV baru
filtered_df.to_csv('filtered_data_mid_level.csv', index=False)
```
