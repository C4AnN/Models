```python
import os
import pandas as pd
import glob
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# Path ke folder utama yang berisi file CSV
base_path = "dataset/January 2021"

# Debug output untuk memastikan path yang digunakan
print(f"Checking files in directory: {base_path}")
print(f"Absolute path: {os.path.abspath(base_path)}")

# Menggunakan glob untuk menemukan semua file CSV di dalam semua sub-folder
all_files = glob.glob(os.path.join(base_path, '**/*.csv'), recursive=True)

# Memeriksa apakah ada file yang ditemukan
if not all_files:
    raise ValueError(f"No CSV files found in the directory and subdirectories: {base_path}")

# Membaca dan menggabungkan semua file CSV
df_list = []
for filename in all_files:
    print(f"Reading file: {filename}")  # Debug output
    try:
        df = pd.read_csv(filename)
        if 'link' in df.columns:
            df = df.drop(columns=['link'])  # Menghapus kolom 'link'
        if not df.empty:
            df_list.append(df)
        else:
            print(f"Warning: {filename} is empty and will be skipped.")
    except Exception as e:
        print(f"Error reading {filename}: {e}")

# Memeriksa apakah ada DataFrame yang valid untuk digabungkan
if not df_list:
    raise ValueError("No valid DataFrames to concatenate. All files may be empty or unreadable.")

# Menggabungkan semua DataFrame dalam daftar menjadi satu DataFrame
combined_df = pd.concat(df_list, ignore_index=True)

# Menyimpan hasil gabungan ke file CSV baru
output_path = "dataset/data_combined_baru.csv"
print(f"Saving combined DataFrame to: {output_path}")  # Debug output
combined_df.to_csv(output_path, index=False)
```

```python
import os
import pandas as pd
from IPython.display import display

# Path ke folder utama yang berisi file CSV
base_path = "dataset/January 2021"

# Menggabungkan data dari semua file CSV di dalam semua sub-folder
combined_df = pd.DataFrame()  # DataFrame kosong untuk menyimpan hasil penggabungan
for root, dirs, files in os.walk(base_path):
    for file in files:
        if file.endswith(".csv"):
            file_path = os.path.join(root, file)
            df = pd.read_csv(file_path)
            if 'link' in df.columns:
                df = df.drop(columns=['link'])  # Menghapus kolom 'link'
            if 'location' in df.columns:
                df = df.drop(columns=['location'])  # Menghapus kolom 'location'
            combined_df = pd.concat([combined_df, df], ignore_index=True)

# Menggabungkan data yang duplikat dan menjumlahkan kolom 'sold'
group_cols = ['name']
if 'sold' in combined_df.columns:
    combined_df['sold'] = pd.to_numeric(combined_df['sold'], errors='coerce')  # Mengonversi kolom 'sold' menjadi tipe numerik
    combined_df.dropna(subset=['sold'], inplace=True)  # Menghapus baris dengan nilai 'sold' yang hilang (NaN)
    combined_df = combined_df.groupby(group_cols, as_index=False).agg({'sold': 'sum', **{col: 'first' for col in combined_df.columns if col not in group_cols + ['sold']}})

# Mengurutkan DataFrame berdasarkan kolom "date"
if 'date' in combined_df.columns:
    combined_df['date'] = pd.to_datetime(combined_df['date'])  # Mengonversi kolom 'date' menjadi tipe datetime
    combined_df = combined_df.sort_values(by='date')  # Mengurutkan DataFrame berdasarkan kolom 'date'

# Menyimpan hasil gabungan ke file CSV baru
output_path = "dataset/data_combined.csv"
combined_df.to_csv(output_path, index=False)
print(f"Saving combined DataFrame to: {output_path}")  # Debug output

# Memuat kembali file CSV gabungan untuk ditampilkan
combined_df_loaded = pd.read_csv(output_path)

# Menampilkan isi DataFrame gabungan
display(combined_df_loaded)
```

```python
import pandas as pd

# Membaca dataset
dataset = pd.read_csv(output_path)

# Membersihkan kolom "price" dari titik sebagai tanda pemisah ribuan
dataset['price'] = dataset['price'].str.replace('.', '')

# Mengonversi kolom "price" ke tipe data integer
dataset['price'] = dataset['price'].astype(int)

# Mengonversi kolom "sold" ke tipe data integer
dataset['sold'] = dataset['sold'].astype(int)

# Memisahkan string tanggal yang dipisahkan oleh koma
dataset['date'] = dataset['date'].str.split(',')

# Mengonversi tanggal dalam kolom "date" menjadi tipe data datetime
dataset['date'] = dataset['date'].apply(lambda x: [pd.to_datetime(date) for date in x])

# Menyimpan dataset yang telah diperbarui
dataset.to_csv("dataset_cleaned.csv", index=False)
```

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM
from tensorflow.keras.layers import Dropout
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt


# 1. Memuat dataset
# Ganti 'path/to/penjualan_laptop.csv' dengan path yang benar ke file CSV Anda
df = pd.read_csv('dataset/dataset_cleaned.csv')

# Pastikan format kolom date benar
df['date'] = pd.to_datetime(df['date'])

# 2. Agregasi data penjualan per minggu
df['week'] = df['date'].dt.to_period('W')
weekly_sales = df.groupby('week').agg({'sold': 'sum'}).reset_index()
weekly_sales['week'] = weekly_sales['week'].dt.to_timestamp()

# Pastikan ada cukup data
num_data_points = len(weekly_sales)
print(f"Jumlah data setelah agregasi mingguan: {num_data_points}")

# Jika data kurang dari 2, kita tidak bisa melanjutkan
if num_data_points < 2:
    raise ValueError("Data tidak cukup untuk membuat prediksi.")

# 3. Skala data
scaler = MinMaxScaler(feature_range=(0, 1))
weekly_sales['sold'] = scaler.fit_transform(weekly_sales[['sold']])

# 4. Mempersiapkan data untuk LSTM
def create_dataset(data, time_step=1):
    X, y = [], []
    for i in range(len(data) - time_step):
        a = data[i:(i + time_step), 0]
        X.append(a)
        y.append(data[i + time_step, 0])
    return np.array(X), np.array(y)

time_step = 1  # Menggunakan 4 minggu sebelumnya untuk memprediksi minggu berikutnya
data = weekly_sales['sold'].values
data = data.reshape(-1, 1)

X, y = create_dataset(data, time_step)

# Cek dimensi data sebelum pembagian
print(f"Shape of X: {X.shape}")
print(f"Shape of y: {y.shape}")

# Pastikan X dan y memiliki data sebelum melanjutkan
if X.shape[0] == 0 or y.shape[0] == 0:
    raise ValueError("Dataset tidak memiliki cukup data setelah diproses dengan create_dataset.")

# Bagi data menjadi train dan test
train_size = int(len(X) * 0.8)
test_size = len(X) - train_size
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Cek dimensi data setelah pembagian
print(f"Shape of X_train: {X_train.shape}")
print(f"Shape of X_test: {X_test.shape}")

# Ubah bentuk input menjadi [samples, time steps, features] untuk LSTM
X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))

# Cek dimensi data setelah reshaping
print(f"Shape of X_train after reshape: {X_train.shape}")
print(f"Shape of X_test after reshape: {X_test.shape}")

# 5. Membuat model LSTM
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(time_step, 1)))
model.add(Dropout(0.2))  # Tambahkan dropout di antara lapisan LSTM
model.add(LSTM(50, return_sequences=False))
model.add(Dropout(0.2))  # Tambahkan dropout lagi sebelum lapisan output
model.add(Dense(1))
model.compile(optimizer='adam', loss='mean_squared_error')

# 6. Melatih model
model.fit(X_train, y_train, epochs=50, batch_size=1, verbose=1)

# 7. Membuat prediksi
train_predict = model.predict(X_train)
test_predict = model.predict(X_test)

# Kembalikan nilai ke skala asli
train_predict = scaler.inverse_transform(train_predict)
test_predict = scaler.inverse_transform(test_predict)
y_train = scaler.inverse_transform([y_train])
y_test = scaler.inverse_transform([y_test])

# Melakukan prediksi pada data test
test_predictions = model.predict(X_test)

# Kembalikan prediksi ke skala aslinya
test_predictions = scaler.inverse_transform(test_predictions)

# Hitung RMSE
rmse = np.sqrt(mean_squared_error(scaler.inverse_transform(y_test.reshape(-1, 1)), test_predictions))
print(f'RMSE: {rmse}')
```

```python
# Plotting hasil prediksi
plt.figure(figsize=(12, 6))
plt.plot(weekly_sales['week'], scaler.inverse_transform(data), label='Data Aktual')

# Plot hasil prediksi data latih
train_predict_plot = np.empty_like(data)
train_predict_plot[:, :] = np.nan
train_predict_plot[time_step:len(train_predict) + time_step, :] = train_predict
plt.plot(weekly_sales['week'][time_step:len(train_predict) + time_step], train_predict_plot[time_step:len(train_predict) + time_step], label='Prediksi Data Latih')

# Plot hasil prediksi data uji
test_predict_plot = np.empty_like(data)
test_predict_plot[:, :] = np.nan
test_predict_plot[len(train_predict) + (time_step * 2) + 1:len(data) - 1, :] = test_predict
plt.plot(weekly_sales['week'][len(train_predict) + (time_step * 2) + 1:len(data) - 1], test_predict_plot[len(train_predict) + (time_step * 2) + 1:len(data) - 1], label='Prediksi Data Uji')

plt.legend()
plt.xlabel('Tanggal')
plt.ylabel('Penjualan')
plt.title('Prediksi Penjualan Laptop Per Minggu')
plt.show()
```
